```python
import cv2
import matplotlib.pyplot as plt
import numpy as np 
image_path = "D:\\Academic\\Semester\\Digital Image Lab\\balloons_noisy.png"
img_bgr = cv2.imread(image_path)
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
bilateral_filtered_bgr = cv2.bilateralFilter(img_bgr, d=9, sigmaColor=75, sigmaSpace=75)
bilateral_filtered_rgb = cv2.cvtColor(bilateral_filtered_bgr, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.imshow(img_rgb) # Display original RGB image
plt.title('Original Image')
plt.axis('off') # Hide axes for cleaner image display
plt.subplot(1, 2, 2)
plt.imshow(bilateral_filtered_rgb) # Display filtered RGB image
plt.title('Bilateral Filtered Image')
plt.axis('off') # Hide axes
plt.tight_layout() # Adjust layout to prevent titles from overlapping
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
def alpha_blend_images(image_path1, image_path2, alpha=0.5):
    img1 = cv2.imread(image_path1)
    img2 = cv2.imread(image_path2)
    if img1 is None:
        print(f"Error: Could not load image from {image_path1}")
        return
    if img2 is None:
        print(f"Error: Could not load image from {image_path2}")
        return
    h1, w1, _ = img1.shape
    img2_resized = cv2.resize(img2, (w1, h1))
    blended_image = cv2.addWeighted(img1, 1 - alpha, img2_resized, alpha, 0)
    img1_rgb = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
    img2_rgb = cv2.cvtColor(img2_resized, cv2.COLOR_BGR2RGB)
    blended_image_rgb = cv2.cvtColor(blended_image, cv2.COLOR_BGR2RGB)
    plt.figure(figsize=(15, 5))
    plt.subplot(1, 3, 1)
    plt.imshow(img1_rgb)
    plt.title('Image 1 (Mickey Mouse)')
    plt.axis('off')

    plt.subplot(1, 3, 2)
    plt.imshow(img2_rgb)
    plt.title('Image 2 (Camera)')
    plt.axis('off')

    plt.subplot(1, 3, 3)
    plt.imshow(blended_image_rgb)
    plt.title(f'Alpha Blended Image (alpha={alpha})')
    plt.axis('off')

    plt.tight_layout()
    plt.show()
camera_image_path = "D:\\Academic\\Semester\\Digital Image Lab\\camera.jpeg"
mickey_image_path = "D:\\Academic\\Semester\\Digital Image Lab\\mickey2.jpg"

alpha_blend_images(mickey_image_path, camera_image_path, alpha=0.6)
```
