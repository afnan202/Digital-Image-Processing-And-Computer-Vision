```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
fig3 = cv2.imread("C:\\Users\\ASUS\\Downloads\\Stairs.bmp", 0)
kernel3 = cv2.getStructuringElement(cv2.MORPH_RECT, (3, 3))
dilated_fig3 = cv2.dilate(fig3, kernel3)
eroded_fig3 = cv2.erode(fig3, kernel3)
morph_gradient = cv2.subtract(dilated_fig3, eroded_fig3)
plt.figure(figsize=(10, 8))
plt.subplot(221), plt.imshow(fig3, cmap='gray'), plt.title('Original Fig3')
plt.subplot(222), plt.imshow(dilated_fig3, cmap='gray'), plt.title('Grayscale Dilation')
plt.subplot(223), plt.imshow(eroded_fig3, cmap='gray'), plt.title('Grayscale Erosion')
plt.subplot(224), plt.imshow(morph_gradient, cmap='gray'), plt.title('Morphological Gradient')
plt.tight_layout()
plt.show()
```
