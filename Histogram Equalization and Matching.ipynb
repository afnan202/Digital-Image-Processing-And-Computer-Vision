```python
import cv2
import matplotlib.pyplot as plt
image = cv2.imread("D:\\Academic\\Semester\\Digital Image Lab\\Fig0310(a)(Moon Phobos).tif", cv2.IMREAD_GRAYSCALE)
equalized = cv2.equalizeHist(image)
plt.figure(figsize=(12, 6))
plt.subplot(2, 2, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.subplot(2, 2, 2)
plt.title("Original Histogram")
plt.hist(image.ravel(), bins=256, range=(0, 256), color='black')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.subplot(2, 2, 3)
plt.title("Equalized Image")
plt.imshow(equalized, cmap='gray')
plt.axis('off')
plt.subplot(2, 2, 4)
plt.title("Equalized Histogram")
plt.hist(equalized.ravel(), bins=256, range=(0, 256), color='black')
plt.xlabel('Pixel Intensity')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

```python
import cv2
import numpy as np
from skimage import exposure
import matplotlib.pyplot as plt
target_image_path = "D:\\Academic\\Semester\\Digital Image Lab\\Fig0310(a)(Moon Phobos).tif"
target_image = cv2.imread(target_image_path, cv2.IMREAD_GRAYSCALE)
reference_image_path = "D:\\Academic\\Semester\\Digital Image Lab\\Southern_Pacific_Shasta_Daylight_Mount_Shasta.JPG"
reference_image = cv2.imread(reference_image_path, cv2.IMREAD_GRAYSCALE)
if target_image is None or reference_image is None:
    print("Error: One or both images could not be loaded. Check the file paths.")
else:
    matched_image = exposure.match_histograms(target_image, reference_image)
    plt.figure(figsize=(15, 10))
    plt.subplot(3, 2, 1)
    plt.title("Target Image")
    plt.imshow(target_image, cmap='gray')
    plt.axis('off')
    plt.subplot(3, 2, 2)
    plt.title("Target Histogram")
    plt.hist(target_image.ravel(), bins=256, range=(0, 256), color='black')
    plt.xlabel('Pixel Intensity')
    plt.ylabel('Frequency')
    plt.subplot(3, 2, 3)
    plt.title("Reference Image")
    plt.imshow(reference_image, cmap='gray')
    plt.axis('off')
    plt.subplot(3, 2, 4)
    plt.title("Reference Histogram")
    plt.hist(reference_image.ravel(), bins=256, range=(0, 256), color='black')
    plt.xlabel('Pixel Intensity')
    plt.ylabel('Frequency')
    plt.subplot(3, 2, 5)
    plt.title("Matched Image")
    plt.imshow(matched_image, cmap='gray')
    plt.axis('off')
    plt.subplot(3, 2, 6)
    plt.title("Matched Histogram")
    plt.hist(matched_image.ravel(), bins=256, range=(0, 256), color='black')
    plt.xlabel('Pixel Intensity')
    plt.ylabel('Frequency')

    plt.tight_layout()
    plt.show()
```
