```python
import cv2
from skimage.exposure import match_histograms
import matplotlib.pyplot as plt
source = cv2.imread("D:\\Academic\\Semester\\Digital Image Lab\\Final2024\\Qs4.jpg")       
reference = cv2.imread("D:\\Academic\\Semester\\Digital Image Lab\\Final2024\\RefQs4.jpg")   
source_rgb = cv2.cvtColor(source, cv2.COLOR_BGR2RGB)
reference_rgb = cv2.cvtColor(reference, cv2.COLOR_BGR2RGB)
matched_rgb = match_histograms(source_rgb, reference_rgb, channel_axis=-1)
matched_bgr = cv2.cvtColor(matched_rgb.astype('uint8'), cv2.COLOR_RGB2BGR)
cv2.imwrite('Qs4_histogram_matched.jpg', matched_bgr)
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
plt.imshow(source_rgb)
plt.title("Source Image")
plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(reference_rgb)
plt.title("Reference Image")
plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(matched_rgb.astype('uint8'))
plt.title("Histogram Matched Output")
plt.axis('off')
plt.tight_layout()
plt.show()

```

```python
#Solution Of Question 02
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread("D:\\Academic\\Semester\\Digital Image Lab\\Final2024\\Qs2.png")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3) 
sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3) 
sobel_combined = cv2.magnitude(sobelx, sobely) 
sobelx = cv2.convertScaleAbs(sobelx)
sobely = cv2.convertScaleAbs(sobely)
sobel_combined = cv2.convertScaleAbs(sobel_combined)
fig, axs = plt.subplots(1, 4, figsize=(20, 5))
axs[0].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
axs[0].set_title("Original Image")
axs[0].axis("off")
axs[1].imshow(sobel_combined, cmap='gray')
axs[1].set_title("Sobel Operator Is Applied")
axs[1].axis("off")
axs[2].imshow(sobelx, cmap='gray')
axs[2].set_title("Sobel X Operator Is Applied. Horizontal Edges")
axs[2].axis("off")
axs[3].imshow(sobely, cmap='gray')
axs[3].set_title("Sobel Y Operator Is Applied. Vertical Edges")
axs[3].axis("off")
plt.tight_layout()
plt.show()
```

```python
#solution of Q3
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread("D:\\Academic\\Semester\\Digital Image Lab\\Final2024\\Qs3.jpg")
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=7)
sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=7)
sobel_combined = cv2.magnitude(sobelx, sobely)
laplacian = cv2.Laplacian(gray, cv2.CV_64F)
canny = cv2.Canny(gray, 100, 200)
titles = ['Original','Sobel Combined', 'Laplacian', 'Canny']
images = [gray,  sobel_combined, laplacian, canny]
plt.figure(figsize=(15, 10))
for i in range(4):
    plt.subplot(2, 3, i+1)
    plt.imshow(np.abs(images[i]), cmap='gray')
    plt.title(titles[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```
